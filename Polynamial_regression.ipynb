```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
#creating dataset
x=10*np.random.normal(0,1,50)
y=10*(x**2)+np.random.normal(-100,100,50)

#Plotting Dataset
plt.scatter(x,y)
plt.xlabel("x random value(Independednt value)")
plt.ylabel("y random value)Dependent Value_")
plt.show()

#linear regression with curveleniar behaviour
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

lm=LinearRegression()
lm.fit(x.reshape(-1,1),y.reshape(-1,1))

y_pred=lm.predict(x.reshape(-1,1))

#plotting linear refression
plt.scatter(x,y,s=10)
plt.plot(x,y_pred,color = 'red')

plt.xlabel("x random value(Independednt value)")
plt.ylabel("y random value)Dependent Value_")

plt.show()

print("Root Mean Squared Error =",np.sqrt(mean_squared_error(y,y_pred)))
print("R2 score =",r2_score(y,y_pred))

#polynomia; regression

input=[('polynomial',PolynomialFeatures(degree=10)),('model',LinearRegression())]
pipe=Pipeline(input)
pipe.fit(x.reshape(-1,1),y.reshape(-1,1))

y_ploy_prod=pipe.predict(x.reshape(-1,1))

#sorting predicted values

sor_zip=sorted(zip(x,y_ploy_prod))
x_poly,y_ploy_prod = zip(*sor_zip)

#plotting polynomial regression
plt.scatter(x,y,s=10)
plt.plot(x_poly,y_ploy_prod,color='orange')
plt.plot(x_poly,y_ploy_prod,color='g',label='Polynomial Regression')


plt.xlabel("x random value(Independednt value)")
plt.ylabel("y random value)Dependent Value_")

plt.show()

print("Root Mean Squared Error =",np.sqrt(mean_squared_error(y,y_ploy_prod)))
print("R2 score",r2_score(y,y_ploy_prod))

```
