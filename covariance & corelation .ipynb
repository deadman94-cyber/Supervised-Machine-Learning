```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x=np.random.rand(50)
y=2*x+np.random.normal(0,0.1,50)

cov_mat=np.cov(x,y)
print('covariance = %.2f'%cov_mat[0,1])
```

```python
x = np.random.rand(50)
y=2*x+np.random.normal(0,0.1,50)

cor_mat=np.corrcoef(x,y)
plt.scatter(x,y)
plt.xlabel('x value')
plt.ylabel('y value')
print('coorelation = %.2f'%cor_mat[0,1])
plt.show()
```
